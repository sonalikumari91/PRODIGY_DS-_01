```python
def find_country_with_highest_population(data):

    highest_population_country = data.sort_values(by='2009', ascending=False).iloc[0]
    return highest_population_country['Country Code'], highest_population_country['2009']


country, population = find_country_with_highest_population(data)

# Print the result
print(f"The country with the highest population in 2009 is '{country}' with a population of {population:.0f}")
```

```python
def find_country_with_minimum_population(data):

    minimum_population_country = data.sort_values(by='2009', ascending=True).iloc[0]
    return minimum_population_country['Country Code'], minimum_population_country['2009']


country, population = find_country_with_minimum_population(data)

# Print the result
print(f"The country with the minimum population in 2009 is '{country}' with a population of {population:.0f}")
```

Now we'll extract zimbabwe population over the years (years and population) 

```python
country = 'Zimbabwe'
country_data_zim = data[data['Country Name'] == country]

```

```python
country_data_zim=country_data_zim.drop({'Country Code','Country Name','Indicator Name', 'Indicator Code'},axis =1).transpose()
```

```python
country_data_zim.columns=['Populatioin']
```

```python
country_data_zim.reset_index(inplace=True)
country_data_zim.columns=['Year','population']
```

```python
country_data_zim.Year=pd.to_datetime(country_data_zim.Year)
```

```python
plt.bar(country_data_zim['Year'].dt.year,country_data_zim['population'],color='pink')
plt.title(f'Population of {country} over Years')
plt.xlabel("Year")
plt.ylabel("population")
plt.grid(axis='y', linestyle='--',alpha=0.7)
plt.xticks(rotation=45)
plt.show()
```

```python
data['Growth Rate'] = data['2023'].pct_change() * 100
```

Growth Rate of ZImbabwe

```python
plt.figure(figsize=(6, 6))
plt.plot(data['2023'], data['Growth Rate'], marker='o')
plt.title(f'Annual Population Growth Rate for {country}')
plt.xlabel('2023')
plt.ylabel('Growth Rate (%)')
plt.grid(True)
plt.show()
```

```python
avg_growth_rate=data.groupby("2023")['Growth Rate'].mean()
plt.plot(avg_growth_rate.index,avg_growth_rate)
plt.title("Average Annual Population Growth Rate")
plt.xlabel("2023")
plt.ylabel("Average Growth Rate (%)")
plt.grid(True)
plt.show()
```
